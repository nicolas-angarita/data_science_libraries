```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.relplot(x = 'total_bill' , y='tip', data=tips, s=100, color ='red')
plt.title('Tips by total bill')

plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col = 'sex', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue = 'size', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', 
            col = 'time', hue = 'size', 
            style = 'smoker', data=tips)
plt.show()
```

```python
sns.histplot(tips.total_bill, kde = True)
plt.show()
```

```python
sns.boxplot(data = tips, x='tip')
plt.show()
```

```python
sns.boxplot(data =tips, y='tip', x='size')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.jointplot(data = tips, x='total_bill', y='tip')
plt.show()
```
