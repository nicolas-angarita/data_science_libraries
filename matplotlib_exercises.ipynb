1. Use matplotlib to plot the following equation:

y = x^2 − x + 2


You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = list(range(-49, 50))
y = [n ** 2 - n + 2 for n in x]

plt.plot(x,y)
plt.title('$y = x^2 − x + 2$')
plt.xlabel('x')
plt.ylabel('y')
plt.annotate(xytext = (0, 500), xy = (0,0), text = 'origin',
             fontsize = 14, c = 'grey', arrowprops = {'width': 1})

plt.grid(ls = ':')
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y =√x

y = x^3

y = 2^x

y = 1/(x+1)

You can use functions from the math module to help implement some of the equations above.


```python
x1 = []
y1 = []
 
for x in range(0, 20):  
    y = np.square(x)
    x1.append(x)
    y1.append(y)    
plt.plot(x1,y1, c = 'red')
plt.grid(ls = ':')
```

```python
x2 = []
y2 = []
 
for x in range(-5, 6):  
    y = x ** 3
    x2.append(x)
    y2.append(y)    
plt.plot(x2,y2, c = 'green')
plt.grid(ls = ':')
```

```python
x3 = []
y3 = []
 
for x in range(-5, 6):  
    y = 2 ** x
    x3.append(x)
    y3.append(y)    
plt.plot(x3,y3, c = 'blue')
plt.grid(True)
```

```python
x4 = []
y4 = []
 
for x in range(0, 10):  
    y = 1/(x+1)
    x4.append(x)
    y4.append(y)    
plt.plot(x4,y4, c = 'grey')
plt.grid()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.


```python
plt.figure(figsize = (10,8))

plt.subplot(221)
plt.plot(x1,y1)
plt.title('y =√x')
plt.grid(ls = ':')


plt.subplot(222)
plt.plot(x2,y2, c='red')
plt.title('$y = x^3$')
plt.grid(ls = ':')


plt.subplot(223)
plt.plot(x3,y3, c='green')
plt.title('$y = 2^x$')
plt.grid(ls = ':')


plt.subplot(224)
plt.plot(x3,y3, c='gold')
plt.title('y = 1/(x+1)')
plt.grid(ls = ':')

plt.subplots_adjust(hspace = .3)
                   

plt.suptitle('Comparison between graphs')
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.


```python
   
plt.plot(x1,y1, c = 'red', label = '$√x$')
plt.plot(x2,y2, c = 'green', label = '$x^3$')   
plt.plot(x3,y3, c = 'blue', label = '$2^x$', alpha = .5)
plt.plot(x4,y4, c = 'grey', label = '$\dfrac{1}{x+1}$')


plt.title('Comparison of equations')
plt.xlabel('x values')
plt.ylabel('y values')
plt.xlim(-4,10)
plt.ylim(-20,20)
plt.legend()
plt.grid(True)
plt.show()
```
