```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(range(1,21))


y = x**2 - x + 2


plt.plot(x,y)
plt.annotate(xytext = (5, 50), xy = (0,0), text = 'origin', fontsize = 14, c = 'dodgerblue')
```
